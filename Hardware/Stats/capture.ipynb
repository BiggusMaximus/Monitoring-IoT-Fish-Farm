```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Acquisition

```python
values = {
    'Analog' : [],
    'pH' : []
}
```

```python
analog_read = float(input("Masukkan data analog : "))
ph_data = float(input("Masukkan data pH : "))

values['Analog'].append(analog_read)
values['pH'].append(ph_data)


df = pd.DataFrame(values)
df.to_csv("./Reading/regression.csv")
df.tail(5)
```

# Data Visualization

```python
df = pd.read_csv('../Reading/regression.csv')
plt.scatter(df['Analog'], df['pH'], facecolors='none', edgecolors='k')
sns.regplot(x="Analog", y="pH", data=df)
```

# Regression Analysis

```python
df = df.sort_values(by=['Analog'])
df = df.reset_index(drop=True)
```

```python
x_analog = df["Analog"].to_numpy().reshape((-1, 1))
y_ph = df["pH"].to_numpy()

model = LinearRegression().fit(x_analog, y_ph)
r_sq = model.score(x_analog, y_ph)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")


y_pred_voltage = model.intercept_ + model.coef_ * x_analog


print(f"{model.intercept_} + {model.coef_[0]} * analog")
plt.scatter(x_analog, y_ph, facecolors='none', edgecolors='k')
plt.xlabel('Analog')
plt.ylabel('Voltage')
plt.plot(x_analog, y_pred_voltage, color='k', linewidth=0.5)
plt.savefig('../Reading/img/regression.png')

plt.show()
```
